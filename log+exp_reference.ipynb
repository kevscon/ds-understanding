# Space

## Log Plot

### Base e

```python
plt.scatter(np.log(np.linspace(0.001, 10, 50)), np.zeros(50))
plt.title('Points in Logscale')
plt.yticks([]);
```

### Base 10

```python
plt.scatter(np.log10(np.linspace(0.001, 10, 50)), np.zeros(50))
plt.title('Points in Logscale')
plt.yticks([]);
```

## Logspace
Return numbers spaced evenly on a log scale with default base 10

```python
# numpy function
np.logspace(0, 2, 20)
```

```python
# ...same as
10**np.linspace(0, 2, 20)
```

### Plot

```python
plt.scatter(np.logspace(0, 2, 20), np.zeros(20))
plt.title('Points in Logspace')
plt.yticks([]);
```

### Plot on Logscale

```python
# logspace points are evenly spaced on a logscale plot
plt.scatter(np.logspace(0, 2, 20), np.zeros(20))
plt.title('Points in Logspace on Logscale')
plt.yticks([])
# base 10
plt.xscale('log');
```

### 2-D Plot

```python
# linear vs logspace plot
plt.scatter(np.linspace(0, 10, 11), np.linspace(0, 10, 11), label='linear-scale')
plt.scatter(np.linspace(0, 10, 11), np.logspace(0, 1, 11), label='log-scale')
plt.title('Logspace Plot')
plt.xlabel('value')
plt.ylabel('scaled value')
plt.legend();
```

## Exponential Plot

```python
plt.scatter(np.exp(np.linspace(0, 10, 50)), np.zeros(50))
plt.title('Exponential Points')
plt.yticks([]);
```

# Transformations

## Logarithmic Transformation

```python
# sample data
x = np.arange(1, 11)
y = np.arange(1, 11)
```

```python
plt.scatter(x, y, label='linear')
plt.scatter(x, np.log(y), label='log(y)')
plt.title('Log Transformation of y')
plt.xlabel('x')
plt.ylabel('y')
plt.legend();
```

```python
fig, ax = plt.subplots()
ax.set_title('Log Transformations')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.scatter(x, y, label='linear')
ax.scatter(np.log(x), y, label='log(x)')
ax.scatter(x, np.log(y), label='log(y)')
ax.scatter(np.log(x), np.log(y), label='log(x), log(y)')
ax.legend();
```

## Exponential Transformation

```python
# sample data
x = np.linspace(0, 5, 10)
y = np.linspace(0, 5, 10)
```

```python
plt.scatter(x, y, label='linear')
plt.scatter(x, np.exp(y), label='exp(y)')
plt.title('Exponential Transformation of y')
plt.xlabel('x')
plt.ylabel('y')
plt.legend();
```

```python
fig, ax = plt.subplots()
ax.set_title('Exponential Transformations')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.scatter(x, y, label='linear')
ax.scatter(np.exp(x), y, label='exp(x)')
ax.scatter(x, np.exp(y), label='exp(y)')
ax.scatter(np.exp(x), np.exp(y), label='exp(x), exp(y)')
ax.legend();
```
