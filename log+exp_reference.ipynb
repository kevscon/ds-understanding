# Logarithmic and Exponential Space <a name="space"></a>

## Log Plot

### Base e

```python
plt.scatter(np.log(np.linspace(0.001, 10, 50)), np.zeros(50))
plt.title('Logarithmic Number Range')
plt.yticks([]);
```

### Base 10

```python
plt.scatter(np.log10(np.linspace(0.001, 10, 50)), np.zeros(50))
plt.title('Logarithmic Number Range')
plt.yticks([]);
```

## Log Space

```python
# return numbers spaced evenly on a log scale with base 10 power input
np.logspace(0, 2, 20)
```

```python
# ...same as
10**np.linspace(0, 2, 20)
```

```python
# return evenly spaced numbers over a specified interval in log scale (base 10)
np.geomspace(1, 100, 20)
```

### Plot

```python
plt.scatter(np.geomspace(1, 100, 20), np.zeros(20))
plt.title('Points in Logscale')
plt.yticks([]);
```

### Log Scale Plot

```python
# logspace points are evenly spaced on a logscale plot
plt.scatter(np.geomspace(1, 100, 20), np.zeros(20))
plt.title('Points in Logscale on Log Scaled Plot')
plt.yticks([])
# base 10
plt.xscale('log');
```

### 2-D Plots

```python
# linear vs logspace plot
plt.scatter(np.linspace(0.1, 10, 11), np.linspace(0.1, 10, 11), label='linear-scale')
plt.scatter(np.linspace(0.1, 10, 11), np.geomspace(0.1, 10, 11), label='log-scale')
plt.title('Logscale Comparison')
plt.xlabel('value')
plt.ylabel('scaled value')
plt.legend();
```

```python
# linear vs logspace plot with y-axis in logscale
plt.scatter(np.linspace(0.1, 10, 11), np.linspace(0.1, 10, 11), label='linear-scale')
plt.scatter(np.linspace(0.1, 10, 11), np.geomspace(0.1, 10, 11), label='log-scale')
plt.title('Logscale Comparison with Logscale Y-axis')
plt.xlabel('value')
plt.ylabel('scaled value')
plt.yscale('log')
plt.legend();
```

## Exponential Plot

```python
plt.scatter(np.exp(np.linspace(0, 10, 50)), np.zeros(50))
plt.title('Exponential Number Range')
plt.yticks([]);
```

## Exponential Space

### Custom Function <a name="exp_func"></a>

```python
def exp_space(start, stop, num_pts=50):
    '''
    Return evenly spaced numbers over a specified interval in an exponential scale.
    
    Parameters
    ----------
    start : starting value of sequence
    stop : end value of sequence
    num_pts : number of points between start and stop to generate
    '''
    
    return(np.log10(np.linspace(10**start, 10**stop, num_pts)))
```

```python
exp_space(0, 5, 20)
```

### Plot

```python
plt.scatter(exp_space(0, 5, 20), np.zeros(20))
plt.title('Points in Exponential Scale')
plt.yticks([]);
```

### 2-D Plots

```python
# linear vs exp_space plot
plt.scatter(np.linspace(0, 10, 11), np.linspace(0, 10, 11), label='linear-scale')
plt.scatter(np.linspace(0, 10, 11), exp_space(0, 10, 11), label='exp-scale')
plt.title('Scale Comparison')
plt.xlabel('value')
plt.ylabel('scaled value')
plt.legend();
```

## Logarithmic and Exponential Plot

```python
# linear vs logspace vs exp_space plot
plt.scatter(np.linspace(0.1, 10, 11), np.linspace(0.1, 10, 11), label='linear-scale')
plt.scatter(np.linspace(0.1, 10, 11), np.geomspace(0.1, 10, 11), label='log-scale')
plt.scatter(np.linspace(0.1, 10, 11), exp_space(0.1, 10, 11), label='exp-scale')
plt.title('Scale Comparison')
plt.xlabel('value')
plt.ylabel('scaled value')
plt.legend();
```

# Transformations <a name="trans"></a>

## Logarithmic Transformation

```python
# sample data
x = np.arange(1, 11)
y = np.arange(1, 11)
```

```python
plt.scatter(x, y, label='linear')
plt.scatter(x, np.log(y), label='log(y)')
plt.title('Log Transformation of y')
plt.xlabel('x')
plt.ylabel('y')
plt.legend();
```

```python
fig, ax = plt.subplots()
ax.set_title('Log Transformations')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.scatter(x, y, label='linear')
ax.scatter(np.log(x), y, label='log(x)')
ax.scatter(x, np.log(y), label='log(y)')
ax.scatter(np.log(x), np.log(y), label='log(x), log(y)')
ax.legend();
```

### Example Distribution
Transform log-normally distributed y_data to log-scale

```python
x_data = np.linspace(1, 1001, 1000)
np.random.seed(0)
# generate random log-normal distributed data
y_data = np.random.lognormal(size=1000)
```

```python
plt.hist(y_data)
plt.title('Histogram');
```

```python
plt.scatter(x_data, y_data)
plt.title('Distribution');
```

```python
# transform to log scale
plt.hist(np.log(y_data))
plt.title('Transformed Histogram');
```

```python
plt.scatter(x_data, np.log(y_data))
plt.title('Transformed Distribution');
```
