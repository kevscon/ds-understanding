Bayesian statistics example for normally distributed data

# Contents

```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

# Data

```python
# population mean height
hgt_avg = 150
# population height standard deviation
hgt_std = 20
# number of samples
hgt_num = 5000
```

## Prior Data

```python
# generate prior data based on population mean and standard deviation
np.random.seed(0)
prior = np.random.normal(loc=hgt_avg, scale=hgt_std, size=hgt_num)
```

```python
# histogram of prior data
plt.hist(prior, bins=50);
```

```python
# prior mean
prior_avg = np.average(prior)
prior_avg
```

```python
# prior standard deviation
prior_std = np.std(prior)
prior_std
```

```python
# minimum sample value
prior_min = min(prior)
prior_min
```

```python
# maximum sample value
prior_max = max(prior)
prior_max
```

```python
# uniform range based on prior values for calcs and plotting
prior_x = np.linspace(prior_min, prior_max, 1000)
```

```python
# calculate normal pdf values based on prior data
prior_pdf = st.norm.pdf(prior_x, loc=prior_avg, scale=prior_std)
```

```python
# plot pdf for example data
plt.plot(prior_x, prior_pdf)
plt.title('Probability Density Function')
plt.xlabel('Observed Value')
plt.ylabel('Probability');
```
