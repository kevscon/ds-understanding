# Imports

```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Generate Data <a name="gen_dat"></a>

## Linear Dataset

```python
x_1 = np.arange(0, 25, 0.25)
# generate variation for dataset
noise_1 = np.random.normal(0, 5, size=len(x_1))
# apply x values and variation to function
y_1 = 2*x_1 + noise_1
```

## Non-linear Dataset

```python
x_2 = np.arange(0, 25, 0.25)
noise_2 =  np.random.normal(0, 5, size=len(x_2))
y_2 = 0.25*x_2**2 + noise_2
```

# Basic Plotting <a name="b_plt"></a>

```python
# line plot
plt.plot(x_1, y_1, x_2, y_2)
```

```python
# scatter plot
plt.scatter(x_1, y_1)
```

```python
# histogram
plt.hist(y_1)
```

```python
# bar plot
plt.bar(x_1[:7], y_1[:7]);
```

## Save Figure

```python
plt.plot(x_1, y_1, x_2, y_2)
plt.savefig('data/plt_ref_fig.png')
```

## Remove Axes Labels

```python
plt.plot(x_1, y_1, x_2, y_2)
# specify no ticks on x axis
plt.xticks([])
# specify no ticks on y axis
plt.yticks([]);
```

## Resize, add title and labels

```python
# change plot size
plt.figure(figsize=(20, 10))
# scatter plot of first dataset, assign color, marker and label
plt.scatter(x_1, y_1, c='b', marker='.', label='first')
# add scatter plot of second dataset
plt.scatter(x_2, y_2, c='r', marker='x', label='second')
# add title to plot
plt.title('Two Datasets')
# add label to x axis
plt.xlabel('x values')
# add label to y axis
plt.ylabel('y values')
# add label legend and specify location
plt.legend(loc='upper left');
```

## Resize markers and text

```python
plt.figure(figsize=(20, 10))
# modify first dataset marker size
plt.scatter(x_1, y_1, c='b', marker='.', label='first', s=150)
# modify second dataset marker size
plt.scatter(x_2, y_2, c='r', marker='x', label='second', s=150)
# modify title font size
plt.title('Two Datasets', fontsize=25)
# modify x axis label font size
plt.xlabel('X values', fontsize=20)
# modify y axis label font size
plt.ylabel('Y values', fontsize=20)
# modify legend font size
plt.legend(loc='upper left', fontsize=15)
# modify axes value fontsize
plt.tick_params(labelsize=15);
```

# Stateful plotting <a name="sta_plt"></a>

## Basic Plot

```python
# define plot object
fig, ax = plt.subplots()
# plot first dataset
ax.scatter(x_1, y_1)
# plot second dataset
ax.scatter(x_2, y_2);
```

## Save Figure

```python
fig.savefig('data/plt_ref_fig.png')
plt.close(fig);
```

## Modify Settings <a name="mod_set"></a>

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax.scatter(x_1, y_1, c='b', marker='.', label='first', s=150)
ax.scatter(x_2, y_2, c='r', marker='x', label='second', s=150)
ax.set_title('Two Datasets', fontsize=25)
ax.set_xlabel('X values', fontsize=20)
ax.set_ylabel('Y values', fontsize=20)
ax.legend(loc='upper left', fontsize=15)
ax.tick_params(labelsize=15);
```

## Multiple Plots <a name="mlt_plt"></a>

```python
# specify plot variable names, number of plots per row and column and overall figure size
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
# first plot
ax1.scatter(x_1, y_1)
# second plot
ax2.scatter(x_2, y_2);
```

### Use list for plots

```python
# specify plot variable list name and overall figure size
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 8))
# first plot
axes[0].scatter(x_1, y_1)
# second plot
axes[1].scatter(x_2, y_2);
```

### Settings for multiple plots

```python
# specify axis to share
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 8), sharex=True)
axes[0].scatter(x_1, y_1)
# remove xticks for first plot
axes[0].set_xticks([])
axes[1].scatter(x_2, y_2)
# set x axis label for second plot
axes[1].set_xlabel('X values')
# set y axis labels for both plots
for ax in axes:
    ax.set_ylabel('Y values')
# set title for entire figure
fig.suptitle('Two Plots');
```

## Axes Ticks <a name="ax_tic"></a>

### Hide ticks and labels

```python
fig, ax = plt.subplots()
ax.scatter(x_1, y_1)
ax.scatter(x_2, y_2)
# set ticks and labels as blank
ax.set_xticks([])
ax.set_yticks([])
ax.set_xticklabels([])
ax.set_yticklabels([]);
```

### Modify ticks and labels

```python
fig, ax = plt.subplots()
ax.scatter(x_1, y_1)
ax.scatter(x_2, y_2)

new_labels = ['Tony', 'Mary', 'Liza', 'Rasputin']
n_ticks = len(new_labels)
start_tick = min(min(x_1), min(x_2))
end_tick = max(max(x_1), max(x_2))

# modify number and spacing of ticks
ax.set_xticks(np.linspace(start_tick, end_tick, n_ticks))
# input tick labels and change orientation
ax.set_xticklabels(new_labels, rotation='vertical');
```

## Axes Limits <a name="ax_lim"></a>

```python
fig, ax = plt.subplots()
ax.scatter(x_1, y_1)
ax.scatter(x_2, y_2)
# set custom limits
ax.set_xlim([5, 15])
ax.set_ylim([0, 75]);
```

## Axes Scale <a name="ax_scl"></a>

### Plot in log scale

```python
fig, ax = plt.subplots()
ax.scatter(x_1, y_1)
ax.scatter(x_2, y_2)
# set y scale
ax.set_yscale('log');
```

## Vertical Line <a name="vert_lin"></a>

```python
fig, ax = plt.subplots()
ax.scatter(x_1, y_1)
ax.scatter(x_2, y_2)
# plot vertical line at specified location
ax.axvline(x=12, linewidth=2, linestyle='dashed', c='g');
```

## Trendlines <a name="trend"></a>

### Fit to linear data

```python
fig, ax = plt.subplots()
ax.scatter(x_1, y_1)
z = np.polyfit(x_1, y_1, 1)
p = np.poly1d(z)
ax.plot(x_1, p(x_1), "r-", linewidth=1);
```

### Fit to non-linear data

```python
fig, ax = plt.subplots()
ax.scatter(x_2, y_2)
z = np.polyfit(x_2, y_2, 2)
p = np.poly1d(z)
ax.plot(x_2, p(x_2), "r-", linewidth=1);
```

## Annotating Text <a name="annot"></a>

```python
fig, ax = plt.subplots()
ax.scatter(x_1, y_1)
ax.scatter(x_2, y_2)
ax.axvline(x=12)
# annotate specified location in plot
ax.annotate('important location', xy=(12, 100), xytext=(3, 75),
            arrowprops=dict(facecolor='red'));
```

## WordCloud <a name="wcld"></a>

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

### Data

```python
# load data
from nltk.corpus import movie_reviews
text_words = movie_reviews.words(movie_reviews.fileids('neg')[0])
```

```python
document = " ".join(text_words)
document
```

### WordCloud image

```python
# create wordcloud object
wc = WordCloud()
wc.generate(document)
```

```python
# plot with imshow
plt.imshow(wc)
plt.axis('off');
```

```python
# change background color
wc = WordCloud(background_color='yellow').generate(document)
# set interpolation parameter to smooth text images
plt.imshow(wc, interpolation='bilinear')
plt.axis('off');
```

### Read in image

```python
from PIL import Image
```

```python
# read in image file
mask = np.array(Image.open('data/usa.png'))
plt.imshow(mask)
plt.axis('off');
```

### Plot wordcloud as image

```python
# apply image mask to wordcloud object
wc = WordCloud(mask=mask).generate(document)
plt.imshow(wc)
plt.axis('off');
```

### Match image coloring

```python
# read in image coloring
image_colors = ImageColorGenerator(mask)
```

```python
wc = WordCloud(mask=mask).generate(document)
# apply image coloring to text rendering
plt.imshow(wc.recolor(color_func=image_colors))
plt.axis('off');
```

### Multiple imshow plots

```python
wc = WordCloud(mask=mask, background_color='white').generate(document)

fig, axes = plt.subplots(1, 3)
axes[0].imshow(mask)
axes[1].imshow(wc)
axes[2].imshow(wc.recolor(color_func=image_colors))
for ax in axes:
    ax.set_axis_off()
```

### Save image

```python
wc.to_file('data/wc_image.png')
```
