# Imports

```python
import random
import numpy as np
import pandas as pd
```

# Generate Data

## Linear Dataset

```python
x_1 = np.arange(0, 25, 0.25)
# generate variation for dataset
noise_1 = np.random.normal(0, 5, size=len(x_1))
# apply x values and variation to function
y_1 = 2*x_1 + noise_1
```

## Non-linear Dataset

```python
x_2 = np.arange(0, 25, 0.25)
noise_2 =  np.random.normal(0, 5, size=len(x_2))
y_2 = 0.25*x_2**2 + noise_2
```

# Stateful plotting <a name="sta_plt"></a>

## Basic Plot

```python
fig, ax = plt.subplots()
ax.scatter(x_1, y_1)
ax.scatter(x_2, y_2);
```

## Save Figure

```python
fig.savefig('data/plt_ref_fig.png')
plt.close(fig);
```

## Modify Settings <a name="mod_set"></a>

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax.scatter(x_1, y_1, c='b', marker='.', label='first', s=150)
ax.scatter(x_2, y_2, c='r', marker='x', label='second', s=150)
ax.set_title('Two Datasets', fontsize=25)
ax.set_xlabel('X values', fontsize=20)
ax.set_ylabel('Y values', fontsize=20)
ax.legend(loc='upper left', fontsize=15)
ax.tick_params(labelsize=15);
```

## Multiple Plots <a name="mlt_plt"></a>

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))
axes[0].scatter(x_1, y_1)
axes[1].scatter(x_2, y_2);
```

## Axes Ticks <a name="ax_tic"></a>

### Hide ticks and labels

```python
fig, ax = plt.subplots()
ax.scatter(x_1, y_1)
ax.scatter(x_2, y_2)
# set ticks and labels as blank
ax.set_xticks([])
ax.set_yticks([])
ax.set_xticklabels([])
ax.set_yticklabels([]);
```

### Modify ticks and labels

```python
fig, ax = plt.subplots()
ax.scatter(x_1, y_1)
ax.scatter(x_2, y_2)

new_labels = ['Tony', 'Mary', 'Liza', 'Rasputin']
n_ticks = len(new_labels)
start_tick = min(min(x_1), min(x_2))
end_tick = max(max(x_1), max(x_2))

# modify number and spacing of ticks
ax.set_xticks(np.linspace(start_tick, end_tick, n_ticks))
# input tick labels and change orientation
ax.set_xticklabels(new_labels, rotation='vertical');
```

## Axes Limits <a name="ax_lim"></a>

```python
fig, ax = plt.subplots()
ax.scatter(x_1, y_1)
ax.scatter(x_2, y_2)
# set custom limits
ax.set_xlim([5, 15])
ax.set_ylim([0, 75]);
```

## Axes Scale <a name="ax_scl"></a>

### Plot in log scale

```python
fig, ax = plt.subplots()
ax.scatter(x_1, y_1)
ax.scatter(x_2, y_2)
# set y scale
ax.set_yscale('log');
```

## Vertical Line <a name="vert_lin"></a>

```python
fig, ax = plt.subplots()
ax.scatter(x_1, y_1)
ax.scatter(x_2, y_2)
# plot vertical line at specified location
ax.axvline(x=12, linewidth=2, linestyle='dashed', c='g');
```

## Trendlines <a name="trend"></a>

### Fit to linear data

```python
fig, ax = plt.subplots()
ax.scatter(x_1, y_1)
z = np.polyfit(x_1, y_1, 1)
p = np.poly1d(z)
ax.plot(x_1, p(x_1), "r-", linewidth=1);
```

### Fit to non-linear data

```python
fig, ax = plt.subplots()
ax.scatter(x_2, y_2)
z = np.polyfit(x_2, y_2, 2)
p = np.poly1d(z)
ax.plot(x_2, p(x_2), "r-", linewidth=1);
```

## Annotating Text <a name="annot"></a>

```python
fig, ax = plt.subplots()
ax.scatter(x_1, y_1)
ax.scatter(x_2, y_2)
ax.axvline(x=12)
# annotate specified location in plot
ax.annotate('important location', xy=(12, 100), xytext=(3, 75),
            arrowprops=dict(facecolor='red'));
```

# Seaborn <a name="sns"></a>

```python
import seaborn as sns
```

## Generate Dataframe Data

```python
# generate dataframe of random integers
df = pd.DataFrame(np.random.randint(0,100,size=(100, 4)), columns=list('ABCD'))
df.head()
```

```python
# generate random string categories for each row
target_set = ['one', 'two', 'three']
target = random.choices(target_set, k=len(df))
```

```python
df['target'] = target
df.head()
```

## lmplot <a name="lmplt"></a>

```python
# scatter plot, x and y are specified dataframe columns
sns.lmplot(x='A', y='B',data=df, fit_reg=False);
```

```python
# scatter plot with regression line
sns.lmplot(x='A', y='B',data=df);
```

```python
# plot with color hue
sns.lmplot(x='A', y='B',data=df, fit_reg=False, hue='target');
```

## Histogram <a name="sns_hist"></a>

```python
sns.distplot(y_1);
```

## Barplot <a name="sns_bar"></a>

```python
sns.countplot(x='A', data=df);
```

## Pairplot <a name="prplt"></a>

```python
sns.pairplot(df);
```

```python
# Plot colored by target column
sns.pairplot(df, 
             vars = ['A', 'B', 'C', 'D'], 
             hue = 'target', diag_kind = 'kde', 
             plot_kws = {'alpha': 0.6, 's': 80, 'edgecolor': 'k'},
             size = 4);
plt.suptitle('Pair Plot of Data', 
             size = 28);
```

## Heat Map <a name="ht_map"></a>

```python
correlation = df.corr()
correlation
```

```python
sns.heatmap(correlation);
```

```python
# reverse color scale
sns.heatmap(correlation, cmap=sns.cm.rocket_r);
```

## KDE Plot <a name="kde"></a>

```python
sns.kdeplot(x_1, y_1);
```

```python
# multiple datasets
ax = sns.kdeplot(x_1, y_1, cmap="Reds")
ax = sns.kdeplot(x_2, y_2, cmap="Blues")
```

```python
# continuous shading
sns.kdeplot(x_1, y_1, n_levels=60, shade=True);
```

```python
# modify color
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.kdeplot(x_1, y_1, cmap=cmap, n_levels=60, shade=True);
```

```python
# jointplot
sns.jointplot(x="A", y="B", data=df, kind="kde");
```

```python
sns.violinplot(data=violin_df['x']);
```

```python
sns.violinplot(data=violin_df);
```
