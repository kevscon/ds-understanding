## Logspace

```python
# logspace output for values between 10^0 and 10^2
np.logspace(0, 2, 11)
```

```python
# equivalent output using linspace
10**np.linspace(0, 2, 11)
```

```python
# linear vs logspace plot
plt.scatter(np.linspace(0, 100, 11), np.linspace(0, 100, 11), label='linear-scale')
plt.scatter(np.linspace(0, 100, 11), np.logspace(0, 2, 11), label='log-scale')
plt.title('Logspace Plot')
plt.xlabel('value')
plt.ylabel('scaled value')
plt.legend();
```

## Spacelog

```python
np.linspace()
```

```python
def space_log(start, stop, num_pts=50):
    '''
    Return evenly spaced numbers over a specified interval in log scale.
    
    Parameters
    ----------
    start : starting value of sequence (must be > 0)
    stop : end value of sequence
    num_pts : number of points between start and stop to generate
    '''
    
    return(np.exp(np.linspace(np.log(start), np.log(stop), num_pts)))
```

```python
# linear vs spacelog plot
plt.scatter(np.linspace(0.001, 100, 11), np.linspace(0.001, 100, 11), label='linear-scale')
plt.scatter(np.linspace(0.001, 100, 11), space_log(0.001, 100, 11), label='log-scale')
plt.title('Spacelog Plot')
plt.xlabel('value')
plt.ylabel('scaled value')
plt.legend();
```

## Log Transform of Axes

```python
# sample data
x = np.arange(1, 11)
y = np.arange(1, 11)
```

```python
plt.scatter(x, y, label='linear')
plt.scatter(x, np.log(y), label='log(y)')
plt.title('Log Transformation of Y-axis')
plt.xlabel('x')
plt.ylabel('y')
plt.legend();
```

```python
fig, ax = plt.subplots()
ax.set_title('Log Transformations')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.scatter(x, y, label='linear')
ax.scatter(np.log(x), y, label='log(x)')
ax.scatter(x, np.log(y), label='log(y)')
ax.scatter(np.log(x), np.log(y), label='log(x), log(y)')
ax.legend();
```
